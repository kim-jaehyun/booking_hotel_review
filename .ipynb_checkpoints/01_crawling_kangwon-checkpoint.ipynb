```python
hotel_name = []
hotel_tourist = []
hotel_review = []
url = 'https://kr.hotels.com/search.do?resolved-location=CITY%3A757364%3AUNKNOWN%3AUNKNOWN&f-star-rating=5,4,3&g-quality-type-id=-2120873974,1052412893&destination-id=757364&q-destination=%EC%9D%B8%EC%B2%9C,%20%EC%9D%B8%EC%B2%9C,%20%ED%95%9C%EA%B5%AD&q-check-in=2020-11-17&q-check-out=2020-11-19&q-rooms=1&q-room-0-adults=2&q-room-0-children=0'

for a in range(1, 3): # 호텔수
    try:
        driver.get(url)
        time.sleep(0.5)

        hotel_name = driver.find_element_by_xpath(
                '//*[@id="listings"]/ol/li[{0}]/article/section/div/h3/a'.format(a)).text #호텔이름 xpath
        time.sleep(0.5) 
        driver.find_element_by_xpath(
            '//*[@id="listings"]/ol/li[{0}]/article/section/div/h3/a'.format(a)).click() # 호텔이름
        print(hotel_name)
        driver.close()
        driver.switch_to.window(driver.window_handles[0])
        time.sleep(1.0)
        print('switch window')
        hotel_tourist = driver.find_element_by_xpath(
            '//*[@id="flexible-container-bottom"]/div[2]/div/div/div[2]').text # 인기명소
        print(hotel_tourist)
        try:
            driver.find_element_by_xpath(
                '//*[@id="property-reviews"]/div[1]/p/a').click() # 이용후기 버튼
            time.sleep(1.0)
        except:
            print('except')
            driver.find_element_by_xpath(
                '//*[@id="property-reviews"]/div[3]/p/a').click() # 이용후기 버튼
            continue
            
            
            
        for i in range(1, 51):
            try:
                hotel_review = driver.find_element_by_xpath(
                    '//*[@id="guest-reviews"]/div[2]/div[1]/div[3]/div[2]/div[{0}]/blockquote'.format(i)).text # 리뷰
                print(hotel_review)
            except:
                continue
        driver.find_element_by_xpath(
            '//*[@id="guest-reviews"]/div[2]/div[1]/div[3]/div[2]/div[51]/div[2]/a').click()
            
        
        time.sleep(0.2)
        hotel_name.append(hotel_name)
        hotel_tourist.append(hotel_tourist)
        hotel_review.append(hotel_review)
    except:
        continue
driver.close()
```
