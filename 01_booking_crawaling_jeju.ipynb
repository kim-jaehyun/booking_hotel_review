```python
hotel_names = []
hotel_tourlists = []
hotel_reviews = []
driver.get(url)

for k in range(1,31): # 호텔페이지
    try:
        driver.find_element_by_xpath('//*[@id="search_results_table"]/div[4]/nav/ul/li[3]/a').click() #호텔 페이지 다음 클릭
        time.sleep(2.0)
    except:
        print('page')
    for i in range (1,29): # 호텔수
        try:
            time.sleep(2.0)
            hotel_name = driver.find_element_by_xpath(
                '//*[@id="hotellist_inner"]/div[{0}]/div[2]/div[1]/div[1]/div[1]/h3/a/span[1]'.format(i)).text #호텔 이름
            print(hotel_name)
            driver.find_element_by_xpath(
            '//*[@id="hotellist_inner"]/div[{0}]/div[2]/div[3]/div/div/div/div/div[1]'.format(i)).click() #호텔 클릭
            time.sleep(3.0)
            driver.switch_to.window(driver.window_handles[1])
            time.sleep(3.0)
            hotel_tourlist = driver.find_element_by_xpath(
                '//div[@class="hp_location_block__content_container hp-poi-content-container--column"]').text #명소 크롤링
            print(hotel_tourlist)
            try:
                driver.find_element_by_xpath('//*[@id="show_reviews_tab"]/span').click() #리뷰 더 보기 클릭
                time.sleep(1.0)
                driver.find_element_by_xpath(
                    '//*[@id="review_list_score_container"]/div[3]/div[1]/div[2]/div/div/label[1]/span').click() #한국어 클릭
            except:
                driver.close()
                driver.switch_to.window(driver.window_handles[0])
            for j in range(1,13): #리뷰 페이지
                try:
                    for y in range(1,7):
                        try:
                            hotel_review = driver.find_element_by_xpath(
                                '//*[@id="review_list_page_container"]/ul/li[{0}]/div/div[4]'.format(y)).text #고객후기
                            print(hotel_review)
                            hotel_names.append(hotel_name)
                            hotel_tourlists.append(hotel_tourlist)
                            hotel_reviews.append(hotel_review)
                        except:
                            continue
                    driver.find_element_by_xpath('//*[@id="review_list_page_container"]/div[4]/div/div[1]/div/div[3]').click() #다음버튼 클릭
                    time.sleep(3.0)
                except:
                    continue
            driver.close()
            driver.switch_to.window(driver.window_handles[0])
    
        except:
            continue
driver.close()
```

```python
df_review = pd.DataFrame({'hotel_name':hotel_names, 'hotel_review':hotel_reviews, 'hotel_tourlist':hotel_tourlists})
df_review.head()
#df_review.tail()
#df_review.info()
```
