```python
import re
import requests
import time
import pandas as pd
import numpy as np
from konlpy.tag import Okt
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium import webdriver
from selenium.webdriver.common.keys import Keys #키워드 검색해야하니까 임포트 해주기 
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import ElementNotInteractableException
```

```python
options = webdriver.ChromeOptions()
#options.add_argument('headless')  # headless를 주면 동작하는 모습을 볼 수 없음
options.add_argument('disable-gpu')
options.add_argument('lang=ko_KR')
driver = webdriver.Chrome('chromedriver', options=options)
driver.implicitly_wait(3)
```

```python
driver = webdriver.Chrome('chromedriver', options=options)
hotel_names = []
hotel_tourlists = []
hotel_reviews = []
url = 'https://kr.hotels.com/search.do?resolved-location=CITY%3A759017%3AUNKNOWN%3AUNKNOWN&f-star-rating=5,4,3&f-accid=3,1&g-quality-type-id=-2120873974,1052412893&g-property-type-id=-1700295402,1621362132&destination-id=759017&q-destination=%EA%B0%95%EB%A6%89,%20%EA%B0%95%EC%9B%90,%20%ED%95%9C%EA%B5%AD&q-check-in=2020-10-02&q-check-out=2020-10-03&q-rooms=1&q-room-0-adults=2&q-room-0-children=0'


driver.get(url)
for i in range(13):
    try:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(5.0)
    except:
        print('scroll')
    for a in range(1, 9):
        try:
            driver.switch_to.window(driver.window_handles[0])
            hotel_name = driver.find_element_by_xpath(
                    '//*[@id="listings"]/ol/li[{0}]/article/section/div/h3/a'.format(i*8+a)).text #호텔이름 xpath
            time.sleep(2.0)
            driver.find_element_by_xpath(
                '//*[@id="listings"]/ol/li[{0}]/article/section/div/h3/a'.format(i*8+a)).click() # 호텔이름
            print(hotel_name)
            driver.switch_to.window(driver.window_handles[1])
            time.sleep(1.0)
            print('switch window')
            hotel_tourlist = driver.find_element_by_xpath(
            '//*[@id="flexible-container-bottom"]/div[2]/div/div/div[2]').text # 인기명소
            print(hotel_tourlist)
            try:
                driver.find_element_by_xpath(
                    '//*[@id="property-reviews"]/div[1]/p/a').click() # 이용후기 버튼
                time.sleep(1.0)
            except:
                print('except')
                driver.find_element_by_xpath(
                    '//*[@id="property-reviews"]/div[3]/p/a').click() # 이용후기 버튼
                continue
        #리뷰갯수
            for i in range(1, 51):
                try:
                    hotel_review = driver.find_element_by_xpath(
                        '//*[@id="guest-reviews"]/div[2]/div[1]/div[3]/div[2]/div[{0}]/blockquote'.format(i)).text # 리뷰
                    print(hotel_review)
                    hotel_names.append(hotel_name)
                    hotel_tourlists.append(hotel_tourlist)
                    hotel_reviews.append(hotel_review)
                    driver.back()
                    time.sleep(0.5)
                except:
                    continue
            driver.find_element_by_xpath(
                        '//*[@id="guest-reviews"]/div[2]/div[1]/div[3]/div[2]/div[51]/div[2]/a').click()
            time.sleep(3.0)
            #다음 버튼 누르는 포문
            for c in range(1, 3):
                try:
                    for d in range(1, 51):
                        try:
                            hotel_review = driver.find_element_by_xpath(
                                '//*[@id="guest-reviews"]/div[2]/div[1]/div[3]/div[2]/div[{0}]/blockquote'.format(d)).text # 리뷰
                            print(hotel_review)
                            hotel_names.append(hotel_name)
                            hotel_tourlists.append(hotel_tourlist)
                            hotel_reviews.append(hotel_review)
                            driver.back()
                            time.sleep(0.5)
                        except:
                            continue
                    driver.find_element_by_xpath(
                        '//*[@id="guest-reviews"]/div[2]/div[1]/div[3]/div[2]/div[51]/div[2]/a[2]').click()
                    time.sleep(3.0)
                except:
                    print('e')
                    driver.find_element_by_xpath(
                        '//*[@id="guest-reviews"]/div[2]/div[1]/div[3]/div[2]/div[51]/div[2]/a[2]').click()
            driver.close()
        except:
            continue
driver.close()
```
